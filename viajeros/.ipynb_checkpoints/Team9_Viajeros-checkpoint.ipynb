## Cambio de Formato y Estandarización

### Conversión a formato 'Long'

```python
def organice(df,column):
    df = df[['Origen','Item',column]]
    df = df.melt(id_vars=['Origen','Item'],value_vars=[column],var_name='Mes', value_name='Viajeros')
    df = df[['Mes','Item','Viajeros','Origen']]
    df['Mes'] = df['Mes'].fillna(value=0)
    df['Mes'] = df['Mes'].str.upper()
    
    return df
```

### Estandarización

```python
def standarize(df):
    #Flag of standarization
    stand = False
    
    # List of Countries
    temp = df['Origen'].unique().tolist()
    temp = sorted(temp)
    
    # List of Items per Sheet
    temp2 = df['Item'].unique().tolist()
    temp2 = sorted(temp2)
    
    # Non-modifying the original dataframe
    df_2 = df.copy()

    # Standarization Process
    for i in temp:
        for j in temp2:
            if not (j in df_2.loc[df_2.Origen.str.contains(i)].Item.tolist()):
                Mes = df_2.loc[df_2.Origen.str.contains(i)].Mes.unique().tolist()[0]
                aux = pd.DataFrame(data = [[Mes,j,0,i]],columns=df_2.columns.tolist())
                df_2 = df_2.append(aux,ignore_index = True)
    
    # Organizing the dataframe
    df_2 = df_2.sort_values(by=['Origen','Item']).reset_index(drop = True)
    
    # Calculating the determing flags to stablish the effectiveness
    occurences = [df_2.loc[df_2.Item.str.contains(k)].count()['Origen'] for k in temp2]
    
    # Determining whether was possible to perform the standarization
    
    if len(set(occurences)) == 1: 
        stand = True
    
    return df_2,stand
    
```

## Concatenación

### Establecimiento por Mes

```python
def concatenation(df,df_Exp,month,flag):
    # Just concatenate those standarized df
    if flag == True:
        # Organice the information
        df_Exp_2 = organice(df,month)
        # Standarize the df
        df_Exp_2,flag = standarize(df_Exp_2)
        # Perform the concatenation
        df_Exp_2 = pd.concat([df_Exp, df_Exp_2])
    
    return df_Exp_2,flag
    
```

```python
ind = df.columns.tolist()[2:]
ind
```

```python
# Process for the first month
month = ind[0]
df_Exp = organice(df,month)
df_Exp,flag = standarize(df_Exp)
```

```python
# Process for the rest of the months
for i in ind[1:]:
    month = i
    df_Exp,flag = concatenation(df,df_Exp,month,flag)
    print('For month ',i,', the process was successfull: ',flag)

df_Exp = df_Exp.reset_index(drop=True)
```
